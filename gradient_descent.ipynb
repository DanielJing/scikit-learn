```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def objective_function(x):
    return (3*x**2)+7
```

```python
def gradient(x):
    return 6*x
```

```python
def gradient_descent(lr, iterations):
   x = 12
   attempts = []
   for _ in range(iterations):
       attempts.append(x)
       x = x - lr * gradient(x)
   return attempts
```

```python
lr = 0.05
iterations = 10
history = gradient_descent(lr, iterations)
```

```python
x = np.linspace(-15, 15, 400)
y = objective_function(x)
plt.figure(figsize = (10, 5))
plt.plot(x, y, label = "Objective Function")
plt.plot(history, [objective_function(x) for x in history], label = 'History')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent')
plt.show()
```
