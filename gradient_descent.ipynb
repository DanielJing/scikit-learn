```python
import sympy as sp

# Define symbolic variables
x = sp.Symbol('x')  

# Define objective function symbolically   
def objective_function(x):
    return (3*x**2)+7

# Take derivative wrt x    
def gradient(x):
    return sp.diff(objective_function(x), x)

# Define gradient descent updating x symbolically
def gradient_descent(lr, iterations, threshold):
   x = sp.Symbol('x', real=True)
   x = 12  
   for i in range(iterations):
       x_old = x
       x = x - lr * gradient(x)  
       if abs(x - x_old) < threshold:
           break
   return x

# Run gradient descent   
x_opt = gradient_descent(0.05, 10, 0.001)

print(f'Optimal x: {x_opt}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def objective_function(x):
    return (3*x**2)+7

def gradient(x):
    return 6*x

def gradient_descent(lr, iterations, threshold):
    x = 12
    attempts = []
    prev_x = None
    for _ in range(iterations):
        attempts.append(x)
        x = x - lr * gradient(x)
        if prev_x is not None and abs(x - prev_x) < threshold:
            break
        prev_x = x
    return attempts

lr = 0.05
iterations = 10
threshold = 0.001
history = gradient_descent(lr, iterations, threshold)

plt.figure(figsize=(10, 5))
plt.plot(history, [objective_function(x) for x in history], label="Objective Function")
plt.plot(history, [gradient(x) for x in history], label="Gradient")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
def objective_function(x):
    return (3*x**2)+7
```

```python
def gradient(x):
    return 6*x
```

```python
def gradient_descent(lr, iterations, threshold):
   x = 12
   attempts = []
   for _ in range(iterations):
       attempts.append(x)
       x = x - lr * gradient(x)
   return attempts
```

```python
lr = 0.05
iterations = 10
threshold = 0.001
history = gradient_descent(lr, iterations, threshold)
```

```python
x = np.linspace(-15, 15, 400)
y = objective_function(x)
plt.figure(figsize = (10, 5))
plt.plot(x, y, label = "Objective Function")
plt.plot(history, [objective_function(x) for x in history], label = 'History')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent')
plt.show()
```
